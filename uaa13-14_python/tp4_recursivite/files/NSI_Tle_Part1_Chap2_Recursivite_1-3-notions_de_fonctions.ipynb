<h2 class="alert alert-info">Exercice 5 : trois programmes pour un même résultat</h2>

On considère la suite $(u_n)$ définies par $u_0=1$ et $u_{n+1}=f(u_n)$ pour tout $n\geq0$, où $f$ est la fonction définie par $f(x)=2x+1$. On a donc :
- $u_0=1$
- $u_1=f(u_0)=2u_0+1=3$
- $u_2=f(u_1)=2u_1+1=7$
- etc.

On désire écrire un programme Python pour calculer le terme de rang $n$ de cette suite.

Pour cela, Charles, Pascal et Mathilde ont écrit les trois programmes suivants :
- Programme de Charles :
```python
def calcul_terme_direct(n):
    u = 1
    for i in range(1, n+1):
        u = 2*u + 1
    return u
```
- Programme de Pascal :
```python
def f(x):
    return 2*x + 1

def calcul_terme_avec_f(n):
    u = 1
    for i in range(1, n+1):
        u = f(u)
    return u
```
- Programme de Mathilde :
```python
def f(x):
    return 2*x + 1

def calcul_mystérieux_avec_f(n):
    if n == 0:
        return 1
    else:
        return f(calcul_mystérieux_avec_f(n-1))
```
