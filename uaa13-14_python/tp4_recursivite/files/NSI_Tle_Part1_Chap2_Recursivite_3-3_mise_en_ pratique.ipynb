2. On propose ci dessous deux algorithmes récursifs pour le calcul du reste et du quotient de la division euclidienne sans utiliser les opérateurs Python : `%` et `//`.

````python
def reste_eucl(a : int, b : int):
    """Calcul récursif du reste euclidien
    de a par b (entiers > 0)"""
    if a < b:
        return a
    return reste_eucl(a-b, b)
````

````python
def quotient_eucl(a : int, b : int):
    """Calcul récursif du quotient
    euclidien de a par b (entiers >0)"""
    if a < b:
        return 0
    return 1 + quotient_eucl(a-b, b)
````

<h2 class="alert alert-info">Exercice 3 : approximation de $\pi$ par la méthode de Nicolas de Cues</h2>

La méthode de Nicolas de Cues (1401-1464) est une méthode algorithmique permettant de déterminer un encadrement de $\pi$.

Soient $(a_n)$ et $(b_n)$ deux suites numériques définies par $a_0=0$, $b_0=\dfrac{1}{4}$ et, pour tout $n\geq0$, par les relations de récurrence
$$a_{n+1}=\dfrac{a_n+b_n}{2}\quad\text{et}\quad b_{n+1}=\sqrt{a_{n+1}b_n}.$$
Alors, on peut montrer que les deux suites $\left(\dfrac{1}{2a_n}\right)$ et $\left(\dfrac{1}{2b_n}\right)$ convergent vers $\pi$.

Pour implémenter cette méthode algorithmique, une idée naturelle est d'utiliser la récursivité. On a alors le code suivant :

````python
from math import sqrt

def a(n):
    if n == 0:
        return 0
    return (a(n-1) + b(n-1))/2

def b(n):
    if n == 0:
        return 1/4
    return sqrt(a(n)*b(n-1))

def approx_pi(n):
    assert isinstance(n, int) and n > 0
    return 1/(2*a(n)), 1/(2*b(n))
````
