<h2 class="alert alert-info">Exercice 2 : recherche du maximum dans un tableau non trié</h2>

On considère un tableau `T` de taille `n` composés d'entiers. Le tableau `T` n'est pas trié. On cherche à déterminer la valeur maximale stockée dans ce tableau en utilisant la méthode &laquo; diviser pour régner &raquo;.

Pour cela, on donne la fonction récursive suivante :

```python
def val_max_tableau(T : list, i : int, j : int) -> int:
    '''
    Renvoie la valeur maximum stock ́ee dans le tableau
    d’entiers T entre les indices i et j (avec i <=j)
    '''
    if i == j:
        return T[i]
    else:
        m = (i+j)//2
        a = val_max_tableau(T, i, m)
        b = val_max_tableau(T, m+1, j)
        return max(a, b)
```

<h2 class="alert alert-info">Exercice 3 : tri fusion</h2>

John von Neumann invente le tri fusion en 1945 : il effectue le tri d'un tableau en se basant sur la méthode &laquo; diviser pour régner &raquo;. Sa complexité est $O(n\log_2(n))$ où $n$ désigne le nombre d'éléments du tableau.

**Partie 1 : principe du tri fusion.**

2. L'algorithme du **tri fusion** fonctionne de la façon suivante : si le tableau n'a qu'un seul élément alors il est déjà trié, sinon on réalise les trois étapes suivantes :
 - diviser : découper le tableau en deux tableaux de même taille (à un élément près) ;
 - régner : trier chacun des sous-tableaux, récursivement (on redécoupe encore les sous-tableaux tant que cela est possible) ;
 - combiner : on fusionne les tableaux triés en comparant le premier élément de chaque sous-tableau (construction récursive).

Une implémentation possible du tri fusion en Python est la suivante :

```python
def fusion(T1 : list, T2 : list) -> list:
    '''
    Fusionne les deux tableaux T1 et T2 déjà triés en un nouveau tableau trié.
    '''
    if T1 == []:
        return T2
    if T2 == []:
        return T1
    if T1[0] <= T2[0]:
        return [T1.pop(0)] + fusion(T1, T2)
    else:
        return [T2.pop(0)] + fusion(T1, T2)

def tri_fusion(T : list) -> list:
    '''
    T est un tableau d'éléments comparables entre eux.
    Renvoie un tableau dans lequel les éléments de T ont été trié par ordre croissant de valeurs (nombres ou caractères).
    '''
    if len(T) <= 1:
        return T
    else:
        m = len(T)//2 # calcul de la position médiane (+1)
        return fusion(tri_fusion(T[:m]), tri_fusion(T[m:]))
```
